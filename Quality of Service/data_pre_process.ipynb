```python
# ============================================
# TELECOM DATASET INITIAL DATA AUDIT
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================
# LOAD DATA
# ============================================

df = pd.read_csv("combined_dataset.csv")

# ============================================
# BASIC OVERVIEW
# ============================================

print("\n========== SHAPE ==========")
print(df.shape)

print("\n========== COLUMN NAMES ==========")
print(df.columns.tolist())

print("\n========== DATA TYPES ==========")
print(df.dtypes)

print("\n========== FIRST 5 ROWS ==========")
print(df.head())

print("\n========== LAST 5 ROWS ==========")
print(df.tail())

# ============================================
# DUPLICATES
# ============================================

print("\n========== DUPLICATES ==========")
print("Duplicate Rows:", df.duplicated().sum())

# ============================================
# MISSING VALUES
# ============================================

print("\n========== MISSING VALUES ==========")

missing = pd.DataFrame({
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (
        df.isnull().sum() / len(df)
    ) * 100
})

missing = missing.sort_values(
    by='Missing_Percentage',
    ascending=False
)

print(missing)

# ============================================
# CHECK FOR " - " VALUES
# ============================================

print("\n========== CHECKING ' - ' VALUES ==========")

for col in df.columns:

    count_dash = (df[col] == " - ").sum()

    if count_dash > 0:
        print(f"{col}: {count_dash}")

# ============================================
# REPLACE " - " WITH NaN
# ============================================

df = df.replace(" - ", np.nan)

# ============================================
# NUMERIC COLUMN CONVERSION
# ============================================

numeric_cols = [
    'Longitude',
    'Latitude',
    'Speed',
    'RSRP',
    'RSRQ',
    'SNR',
    'CQI',
    'RSSI',
    'DL_bitrate',
    'UL_bitrate',
    'NRxRSRP',
    'NRxRSRQ',
    'ServingCell_Lon',
    'ServingCell_Lat',
    'ServingCell_Distance'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# ============================================
# DATA TYPES AFTER CONVERSION
# ============================================

print("\n========== DATA TYPES AFTER FIX ==========")
print(df.dtypes)

# ============================================
# DESCRIPTIVE STATISTICS
# ============================================

print("\n========== NUMERIC SUMMARY ==========")
print(df.describe())

# ============================================
# UNIQUE VALUES IN CATEGORICAL COLUMNS
# ============================================

print("\n========== CATEGORICAL UNIQUE VALUES ==========")

cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:

    print(f"\n--- {col} ---")
    print("Unique Count:", df[col].nunique())

    print(df[col].value_counts().head(10))

# ============================================
# CHECK INVALID SIGNAL VALUES
# ============================================

print("\n========== INVALID SIGNAL VALUE CHECK ==========")

checks = {
    'RSRP': (-140, -40),
    'RSRQ': (-25, 0),
    'SNR': (-20, 40),
    'CQI': (0, 15),
    'RSSI': (-120, -30),
    'Speed': (0, 300)
}

for col, (low, high) in checks.items():

    invalid = df[
        (df[col] < low) |
        (df[col] > high)
    ]

    print(f"\n{col}")
    print("Invalid Rows:", len(invalid))

# ============================================
# CORRELATION MATRIX
# ============================================

print("\n========== CORRELATION MATRIX ==========")

corr = df[numeric_cols].corr()

print(corr)

# ============================================
# CORRELATION HEATMAP
# ============================================

plt.figure(figsize=(14,10))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()

# ============================================
# MISSING VALUE VISUALIZATION
# ============================================

plt.figure(figsize=(14,6))

sns.heatmap(
    df.isnull(),
    cbar=False,
    yticklabels=False
)

plt.title("Missing Value Heatmap")
plt.show()

# ============================================
# DISTRIBUTION OF IMPORTANT SIGNAL FEATURES
# ============================================

important_cols = [
    'RSRP',
    'RSRQ',
    'SNR',
    'CQI',
    'RSSI',
    'DL_bitrate',
    'UL_bitrate',
    'ServingCell_Distance'
]

for col in important_cols:

    plt.figure(figsize=(8,4))

    sns.histplot(
        df[col],
        bins=50,
        kde=True
    )

    plt.title(f"Distribution of {col}")
    plt.show()

# ============================================
# OUTLIER CHECK USING BOXPLOTS
# ============================================

for col in important_cols:

    plt.figure(figsize=(8,3))

    sns.boxplot(x=df[col])

    plt.title(f"Boxplot of {col}")
    plt.show()

# ============================================
# NETWORK MODE ANALYSIS
# ============================================

print("\n========== NETWORK MODE DISTRIBUTION ==========")

print(df['NetworkMode'].value_counts())

# ============================================
# OPERATOR ANALYSIS
# ============================================

print("\n========== OPERATOR DISTRIBUTION ==========")

print(df['Operatorname'].value_counts())

# ============================================
# TRANSPORT MODE ANALYSIS
# ============================================

print("\n========== TRANSPORT MODE DISTRIBUTION ==========")

print(df['Transport_Mode'].value_counts())

# ============================================
# CHECK RELATIONSHIP BETWEEN SIGNALS & SPEED
# ============================================

signal_cols = ['RSRP', 'RSRQ', 'SNR', 'CQI']

for col in signal_cols:

    plt.figure(figsize=(7,5))

    sns.scatterplot(
        x=df['Speed'],
        y=df[col],
        alpha=0.3
    )

    plt.title(f"Speed vs {col}")
    plt.show()

# ============================================
# CHECK BITRATE RELATIONSHIPS
# ============================================

plt.figure(figsize=(7,5))

sns.scatterplot(
    x=df['SNR'],
    y=df['DL_bitrate'],
    alpha=0.3
)

plt.title("SNR vs Download Bitrate")
plt.show()

plt.figure(figsize=(7,5))

sns.scatterplot(
    x=df['CQI'],
    y=df['DL_bitrate'],
    alpha=0.3
)

plt.title("CQI vs Download Bitrate")
plt.show()

# ============================================
# CHECK MISSING VALUE CORRELATION
# ============================================

print("\n========== MISSING VALUE CORRELATION ==========")

missing_cols = [
    'SNR',
    'CQI',
    'RSSI',
    'NRxRSRP',
    'NRxRSRQ',
    'ServingCell_Distance'
]

for col in missing_cols:

    print(f"\n{col} missing by NetworkMode:")

    print(
        df.groupby('NetworkMode')[col]
        .apply(lambda x: x.isnull().mean())
        .sort_values(ascending=False)
    )

# ============================================
# FINAL REPORT
# ============================================

print("\n========== FINAL DATA AUDIT COMPLETE ==========")

print("""
Things to observe carefully:

1. Invalid signal ranges
2. Missing value patterns
3. Correlation between telecom metrics
4. Extreme outliers in bitrate
5. NetworkMode distribution
6. Operator imbalance
7. Signal vs Speed relationships
8. Which columns need encoding
9. Which columns may leak QoS target
10. Feature importance intuition
""")
```

```python
# =========================================================
# TELECOM LTE QoS DATASET - COMPLETE CLEANING PIPELINE
# =========================================================

# =========================================================
# IMPORTS
# =========================================================

import pandas as pd
import numpy as np

# =========================================================
# LOAD DATA
# =========================================================

df = pd.read_csv(
    "combined_dataset.csv",
    low_memory=False
)

print("\n================ ORIGINAL DATASET ================")
print("Shape:", df.shape)

# =========================================================
# REMOVE DUPLICATES
# =========================================================

duplicates = df.duplicated().sum()

print("\nDuplicate Rows:", duplicates)

df = df.drop_duplicates()

print("Shape After Removing Duplicates:", df.shape)

# =========================================================
# REPLACE " - " WITH NaN
# =========================================================

df = df.replace(" - ", np.nan)

print("\nReplaced ' - ' with NaN")

# =========================================================
# CONVERT TIMESTAMP
# =========================================================

df['Timestamp'] = pd.to_datetime(
    df['Timestamp'],
    format='%Y.%m.%d_%H.%M.%S',
    errors='coerce'
)

# =========================================================
# NUMERIC COLUMN CONVERSION
# =========================================================

numeric_cols = [
    'Longitude',
    'Latitude',
    'Speed',
    'RSRP',
    'RSRQ',
    'SNR',
    'CQI',
    'RSSI',
    'DL_bitrate',
    'UL_bitrate',
    'NRxRSRP',
    'NRxRSRQ',
    'ServingCell_Lon',
    'ServingCell_Lat',
    'ServingCell_Distance'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )

print("\nNumeric Conversion Completed")

# =========================================================
# KEEP ONLY LTE ROWS
# =========================================================

df = df[
    df['NetworkMode'] == 'LTE'
]

print("\nOnly LTE Rows Kept")
print("Shape:", df.shape)

# =========================================================
# CLEAN OPERATOR NAMES
# =========================================================

invalid_ops = [
    '0',
    '27202',
    '27205',
    '27201',
    '27203'
]

df['Operatorname'] = df['Operatorname'].replace(
    invalid_ops,
    'Unknown'
)

# =========================================================
# REMOVE INVALID SIGNAL VALUES
# =========================================================

# -------------------------
# RSRP
# -------------------------

df = df[
    (df['RSRP'] >= -140) &
    (df['RSRP'] <= -40)
]

# -------------------------
# RSRQ
# -------------------------

df.loc[
    (df['RSRQ'] < -25) |
    (df['RSRQ'] > 0),
    'RSRQ'
] = np.nan

# -------------------------
# SNR
# -------------------------

df.loc[
    (df['SNR'] < -20) |
    (df['SNR'] > 40),
    'SNR'
] = np.nan

# -------------------------
# CQI
# -------------------------

df.loc[
    (df['CQI'] < 0) |
    (df['CQI'] > 15),
    'CQI'
] = np.nan

# -------------------------
# RSSI
# -------------------------

df.loc[
    (df['RSSI'] < -120) |
    (df['RSSI'] > -30),
    'RSSI'
] = np.nan

# -------------------------
# NRxRSRP
# -------------------------

df.loc[
    (df['NRxRSRP'] < -140) |
    (df['NRxRSRP'] > 0),
    'NRxRSRP'
] = np.nan

# -------------------------
# NRxRSRQ
# -------------------------

df.loc[
    (df['NRxRSRQ'] < -40) |
    (df['NRxRSRQ'] > 0),
    'NRxRSRQ'
] = np.nan

# -------------------------
# ServingCell_Distance
# -------------------------

df.loc[
    (df['ServingCell_Distance'] < 0) |
    (df['ServingCell_Distance'] > 50000),
    'ServingCell_Distance'
] = np.nan

# =========================================================
# CREATE MISSING VALUE INDICATORS
# =========================================================

important_missing_cols = [
    'SNR',
    'CQI',
    'RSSI',
    'NRxRSRP',
    'NRxRSRQ',
    'ServingCell_Distance'
]

for col in important_missing_cols:

    df[col + '_missing'] = (
        df[col].isnull().astype(int)
    )

print("\nMissing Indicators Created")

# =========================================================
# IMPUTE MISSING VALUES
# =========================================================

# LTE rows now remain,
# so median imputation is conceptually valid.

impute_cols = [
    'RSRQ',
    'SNR',
    'CQI',
    'RSSI',
    'NRxRSRP',
    'NRxRSRQ',
    'ServingCell_Lon',
    'ServingCell_Lat',
    'ServingCell_Distance'
]

for col in impute_cols:

    median_value = df[col].median()

    df[col] = df[col].fillna(
        median_value
    )

print("\nMedian Imputation Completed")

# =========================================================
# FEATURE ENGINEERING - TIMESTAMP
# =========================================================

df['Hour'] = df['Timestamp'].dt.hour

df['DayOfWeek'] = df['Timestamp'].dt.dayofweek

df['Month'] = df['Timestamp'].dt.month

df['IsWeekend'] = (
    df['DayOfWeek'].isin([5, 6]).astype(int)
)

# =========================================================
# OPTIONAL: DROP TIMESTAMP
# =========================================================

df = df.drop(columns=['Timestamp'])

# =========================================================
# OPTIONAL: DROP SOURCE FILE
# =========================================================

df = df.drop(columns=['Source_File'])

# =========================================================
# OPTIONAL: DROP RSSI
# HIGHLY CORRELATED WITH RSRP
# =========================================================

# Uncomment if needed later

# df = df.drop(columns=['RSSI'])

# =========================================================
# CREATE QoS TARGET
# =========================================================

def create_qos(row):

    # GOOD QoS

    if (
        row['RSRP'] > -90 and
        row['SNR'] > 20 and
        row['DL_bitrate'] > 10000
    ):
        return 2

    # MODERATE QoS

    elif (
        row['RSRP'] > -105 and
        row['SNR'] > 5 and
        row['DL_bitrate'] > 2000
    ):
        return 1

    # POOR QoS

    else:
        return 0

df['QoS'] = df.apply(
    create_qos,
    axis=1
)

print("\nQoS Target Created")

# =========================================================
# CHECK CLASS BALANCE
# =========================================================

print("\n================ QoS DISTRIBUTION ================")

print(
    df['QoS'].value_counts()
)

print("\nQoS Percentage:")

print(
    df['QoS'].value_counts(normalize=True) * 100
)

# =========================================================
# ENCODE CATEGORICAL COLUMNS
# =========================================================

from sklearn.preprocessing import LabelEncoder

cat_cols = [
    'Operatorname',
    'NetworkMode',
    'State',
    'Transport_Mode'
]

label_encoders = {}

for col in cat_cols:

    le = LabelEncoder()

    df[col] = le.fit_transform(
        df[col].astype(str)
    )

    label_encoders[col] = le

print("\nCategorical Encoding Completed")

# =========================================================
# FINAL CHECK
# =========================================================

print("\n================ FINAL DATASET ================")

print("Final Shape:", df.shape)

print("\nMissing Values Remaining:")

print(df.isnull().sum())

print("\nData Types:")

print(df.dtypes)

print("\nQoS Distribution:")

print(df['QoS'].value_counts())

# =========================================================
# SAVE CLEANED DATASET
# =========================================================

df.to_csv(
    "cleaned_lte_qos_dataset.csv",
    index=False
)

print("\n================================================")
print("CLEANED DATASET SAVED SUCCESSFULLY")
print("File: cleaned_lte_qos_dataset.csv")
print("================================================")
```
