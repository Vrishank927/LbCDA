```python
import pandas as pd
df = pd.read_csv(r'C:\Users\vrish\OneDrive\Desktop\LbCDA\data.csv')
```

```python
import numpy as np
df.replace(-1, np.nan, inplace=True)
```

```python

df["Location_Missing"] = (
    df["Latitude"].isnull() |
    df["Longitude"].isnull()
).astype(int)

df["Latitude"] = df["Latitude"].fillna(-999)

df["Longitude"] = df["Longitude"].fillna(-999)
```

```python
df["Location_Missing"] = 0

df.loc[df["State Name"] == "Unknown", "Location_Missing"] = 1
```

```python
print(df.columns)
```

```python
df=pd.get_dummies(df,
    columns=['Operator'],dtype=int)
```

```python
operator_cols = [
    "Operator_Airtel",
    "Operator_BSNL",
    "Operator_Idea",
    "Operator_MTNL",
    "Operator_Other",
    "Operator_RJio",
    "Operator_Telenor",
    "Operator_Vodafone"
]

df[operator_cols] = df[operator_cols].astype(int)
```

```python
df = pd.get_dummies(df, columns=["State Name"], prefix="State")
```

```python
state_cols = [
    "State_Punjab",
    "State_Rajasthan",
    "State_Sikkim",
    "State_Tamil Nadu",
    "State_Telangana",
    "State_Tripura",
    "State_Uttar Pradesh",
    "State_Uttarakhand",
    "State_West Bengal",
    #"State_Unknown"
]

df[state_cols] = df[state_cols].astype(int)
```

```python
df = pd.get_dummies(
    df,
    columns=[
        "In Out Travelling",
        "Network Type",
    ],
    dtype=int
)
```
